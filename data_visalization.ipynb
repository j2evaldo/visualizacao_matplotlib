## **DataSet Irisi**

```python
iris = pd.read_csv('iris.csv')
```

```python
iris.head()
```

```python
fig = plt.figure(figsize = (10,4))
eixo = fig.add_axes([0,0,1,1])

eixo.scatter(iris['comprimento_sépala'], iris['largura_sépala'])
eixo.set_title('Gŕafico de Dispersão', fontsize = 25, y =1.01, weight = 'bold', loc= 'left')
eixo.set_xlabel('Comprimento da Sépala', fontsize = 15)
eixo.set_ylabel('Largura da Sépala', fontsize = 15);
```

### **Colorindo por Grupo**

```python
iris.columns
```

```python
iris['espécie'].unique().tolist()
```

```python
fig = plt.figure(figsize = (10,5))
eixo = fig.add_axes([0,0,1,1])

cores = {'Iris-setosa':'r', 'Iris-versicolor': 'b', 'Iris-virginica': 'g' }

for i in iris['espécie'].unique():
    tmp = iris[iris['espécie'] == i]
    eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'], color = cores[i])


eixo.set_title('Gŕafico de Dispersão', fontsize = 25, y =1.01, weight = 'bold', loc= 'left')
eixo.set_xlabel('Comprimento da Sépala', fontsize = 15)
eixo.set_ylabel('Largura da Sépala', fontsize = 15);
eixo.legend(cores, fontsize = 15)
```

### **Marcadores**

Bom para identificar sobreposição de conteudo

```python
fig = plt.figure(figsize = (10,5))
eixo = fig.add_axes([0,0,1,1])

cores = {'Iris-setosa':'r', 'Iris-versicolor': 'b', 'Iris-virginica': 'g' }
marcadores = {'Iris-setosa':'x', 'Iris-versicolor': 'o', 'Iris-virginica': 'v' }

# Parametro s altera o tamanho do marcador do gráfico
for i in iris['espécie'].unique():
    tmp = iris[iris['espécie'] == i]
    eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'], color = cores[i], marker= marcadores[i], s = 100)



eixo.set_title('Gŕafico de Dispersão', fontsize = 25, y =1.01, weight = 'bold', loc= 'left')
eixo.set_xlabel('Comprimento da Sépala', fontsize = 15)
eixo.set_ylabel('Largura da Sépala', fontsize = 15);
eixo.legend(cores, fontsize = 15);
```

## **Gráfico de Dispersão**

```python
fig = plt.figure(figsize = (10,4))
eixo = fig.add_axes([0,0,1,1])

# Calculando valores que serão exibidos no Gráfico

mu, sigma = iris['comprimento_pétala'].mean(), iris['comprimento_pétala'].std()

eixo.hist(iris['comprimento_pétala'], bins = 20)
eixo.set_title('Histograma', fontsize = 25, y= 1.02)
eixo.set_xlabel('Comprimento da Pétala', fontsize = 15)
#eixo.grid(True)

#  Exibindo dados estatisticos no gráfico
eixo.annotate(f'mu: {mu:.2f} \n sigma: {sigma:.2f}', xy= (4.5, 20), fontsize = 30)

```

```python
fig = plt.figure(figsize = (10,4))
eixo = fig.add_axes([0,0,1,1])

# Calculando valores que serão exibidos no Gráfico

mu, sigma = iris['comprimento_pétala'].mean(), iris['comprimento_pétala'].std()
mediana = iris['comprimento_pétala'].median()

eixo.hist(iris['comprimento_pétala'], bins = 20)
eixo.set_title('Histograma', fontsize = 25, y= 1.02)
eixo.set_xlabel('Comprimento da Pétala', fontsize = 15)
#eixo.grid(True)

#  Exibindo dados estatisticos no gráfico
eixo.annotate(f'mu: {mu:.2f} \n sigma: {sigma:.2f}', xy= (4.5, 20), fontsize = 30)

# Linhas Veticais no Gráfico
eixo.axvline(mu, color = 'r', linestyle = '--')
eixo.annotate('média', xy = (mu -.5, 28), fontsize = 14)

eixo.axvline(mediana, color = 'g', linestyle = '--')
eixo.annotate('mediana', xy = (mediana +.1, 32), fontsize = 14)
```
